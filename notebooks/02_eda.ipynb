## Visualizations

Let's display the plots we generated:

```python
from IPython.display import Image, display
import os

plots_dir = '../assets/plots/'
for plot in sorted(os.listdir(plots_dir)):
    if plot.endswith('.png'):
        print(f"### {plot.replace('_', ' ').replace('.png', '').title()} ###")
        display(Image(filename=os.path.join(plots_dir, plot)))
        print('\n' + '='*50 + '\n')
```
